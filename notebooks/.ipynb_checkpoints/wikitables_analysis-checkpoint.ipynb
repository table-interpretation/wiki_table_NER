### Analysis of the WikiTables dataset

- number of columns
- number of rows
- number of cells
- number of links per cell

```python
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
import numpy as np
```

```python
file1 = open('datasets/wikidata_limaye/wiki_links_random/wiki_links-random.txt', 'r')
Lines = file1.readlines()

wiki_links_ids=[] 
for line in Lines:
    ids = line.split()
    wiki_links_ids.append([int(i) for i in ids])
print(len(wiki_links_ids))
```

```python

numcols1=[]
numdatarows1=[]

numdatacells1=[]
numlinks_table1 = []
avg_numlinks_per_cell1 = []
avg_numlinks_per_linked_cell1 = []

numlinks_cell1 = []


empty_cells1 = []

with tqdm(total=len(wiki_links_ids)) as progress:

    with pd.read_json('datasets/wikidata_limaye/wiki_links_random/wiki_links_random_tables.json',lines=True,chunksize=1) as reader:
        
        for chunk in reader:
            empty_cell_pertable=0
            linkspercell=[]
            links_per_linkedcell=[]
        
            numcols1.append(int(chunk["numCols"].values))
            numdatarows1.append(int(chunk["numDataRows"].values))

            
            table = chunk["tableData"].values[0]
            
            rows = [ r for r in table ]

            for row in rows:
                for cell in row:
                    if cell["text"] == '':
                        empty_cell_pertable+=1
                    soup = BeautifulSoup(cell["tdHtmlString"], "html.parser")
                    atags = soup.find_all("a", href=True)
                    linkspercell.append(len(atags))

                    if len(atags) >= 1:
                        links_per_linkedcell.append(len(atags))
            
            numdatacells1.append(len([cell for row in rows for cell in row]))
            
            numlinks_table1.append(np.sum(linkspercell))
            numlinks_cell1.extend(linkspercell)

            if len(linkspercell)==0:
                avg_numlinks_per_cell1.append(0)
            else:
                avg_numlinks_per_cell1.append(np.mean(linkspercell))

            if len(links_per_linkedcell)==0:
                avg_numlinks_per_linked_cell1.append(0)
            else:
                avg_numlinks_per_linked_cell1.append(np.mean(links_per_linkedcell))

            empty_cells1.append(empty_cell_pertable)

            
            # Analysis on header cells ? header_cells=[c for c in table["tableHeaders"]]
        

            progress.update()
```
