```python
def get_correct_anno(tab_idx):
    
    table = ner_tables[tab_idx][0]
    tab_id = table[0]
    tableHeaders = table[4]
    table_data = table[5]
    row_labels = table[6]
    ser_correct_annot = []
    
    print(tab_id)
    
    table_csv, table_matrix = tab_to_csv(tableHeaders, table_data, sep = '|')

    for i in range(len(row_labels)):
        if len(row_labels[i][0]) > 0:
            ser_correct_annot.append(row_labels[i][0])
    
    outputs = []
    for anno in ser_correct_annot:   
        for a in anno:
            if a[-1] != 0:
               # print(a)
                cell_index = [a[0], a[1]]
                type_ = labels_dict_rev[a[-1]]
                text = table_matrix[a[0]][a[1]][a[2]:a[3]]                   

                outputs.append({"entity":text, "type":type_, "cell_index":cell_index})
                
                                                    
    return table_csv, table_matrix, outputs
```

```python
def get_labels(test): 
    incorrect_span_predict = []
    incorrect_span_gt = []
    true_label = []
    predicted_label = []
    for key, value in test.items():    
        for idx in range(len(ner_tables)):    

            ser_correct_annot = []        
            table = ner_tables[idx][0]
            tab_id = table[0]

            if tab_id == key:

                tab_csv, tab_matrix, outs = get_correct_anno(idx)            
                predicted = gpt_responce(tab_matrix, test[key], labels_dict)

                row_labels = table[6]
                for i in range(len(row_labels)):
                    ser_correct_annot.append([el for el in row_labels[i][0]])

                ser_gt = [tuple(item) for sublist in ser_correct_annot for item in sublist]
                
                
                for el in predicted:
                    for tup in ser_gt:
                        if el[0:4] == tup[0:4]:
                            true_label.append(tup[4])
                            predicted_label.append(el[4])
                        elif el[0:3] == tup[0:3] and el[4]==tup[4]:
                            incorrect_span_predict.append(el)
                            incorrect_span_gt.append(tup)
    return true_label, predicted_label, incorrect_span_predict, incorrect_span_gt
```
